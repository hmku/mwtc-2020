## Setup

```python
import numpy as np
import pandas as pd
from statistics import stdev
import math
```

```python
# risk free rate, monthly increments (21 days a month)
dfr = pd.read_csv("risk_free_train.csv", index_col=0)
dfr
```

```python
# stock prices, daily increments (252 days a year)
txs = list(map(str, list(range(10)))) + ["market_index"] # useful to have
dfp = pd.read_csv("stock_prices_train.csv", index_col=0)
dfp
```

```python
# helper function for plotting
def plot(df, *args, cs=False):
    if len(args) == 0:
        args = list(df.columns)
    args = list(map(str, args))
    dft = pd.DataFrame()
    for arg in args:
        dft[arg] = df[arg].cumsum() if cs else df[arg]
    dft.plot(figsize=(15, 10), y=args)
```

```python
plot(dfr)
```

```python
plot(dfp)
```

## Returns

```python
# get log returns
dfpr = pd.DataFrame()
for tx in txs:
    dfpr[tx] = np.log(1 + dfp[tx].pct_change())
dfpr
```

```python
# plot cumulative log returns
plot(dfpr, cs=True)
```

## Regression

```python
# calculate covariance matrix, correlations, betas
# beta(X, Y) = cov(X, Y) / var(Y)
# corr(X, Y) = cov(X, Y) / (sd(X)*sd(Y))
covs = dfpr.cov()
vs = dfpr.var()
betas = covs / vs
sds = pd.DataFrame(np.sqrt(vs)) # sds of tickers, but converted into a dataframe
sdp = sds.dot(sds.T) # sdp[X, Y] = sd(X) * sd(Y)
corrs = covs / sdp
```

```python
# correlations
corrs
```

```python
# betas
betas
```

```python
plot(corrs, cs=True)
```

```python
plot(betas, cs=True)
```

```python
# Measure daily Sharpe ratio
def score(weights, stock_prices, market_price, risk_free_rate, daily_excess_return):
    returns = 0
    
    # Calculate daily excess return
    for i in range(len(stock_prices)):
        returns += weights[i] * stock_prices[i]
    
    der = returns - market_price - risk_free_rate
    
    # Track daily excess returns
    daily_excess_return.append(der)
    
    # Calculate Sharpe ratio for one day
    return (der/stdev(daily_excess_return)) * math.sqrt(252)
```
